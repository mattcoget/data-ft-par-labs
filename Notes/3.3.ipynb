```python
url='https://www.reuters.com/article/us-health-coronavirus-finance-breakingvi/breakingviews-corona-capital-tiffany-biden-vaccines-cash-box-idUSKBN2842KL'
```

```python
resp=r.get(url)
```

```python
soup=BeautifulSoup(resp.content)
```

```python
soup.select('p,h1,h2')
```

```python
lot=soup.find_all(['h1','h2','h3','h4','h5','h6','p'])
```

```python
print("\n\t".join([i.text for i in lot]))
```
